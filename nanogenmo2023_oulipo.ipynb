```python
poem_lines = [item.strip() for item in poem.splitlines()]
```

```python
import re

def get_alphanumeric_parts(s):
  return re.findall(r'\w+', s)

def modify_alphanumeric_parts(parts):
  return ['[MASK]']#[part.upper() for part in parts]

def replace_alphanumeric_parts(s, modified_parts):
  for original_part, modified_part in zip(get_alphanumeric_parts(s), modified_parts):
      s = re.sub(original_part, modified_part, s)
  return s

def modify_string(s):
  alphanumeric_parts = get_alphanumeric_parts(s)
  modified_parts = modify_alphanumeric_parts(alphanumeric_parts)
  return replace_alphanumeric_parts(s, modified_parts)
```

```python
with open('poems_new.txt','a') as f:
    total_words = 0
    counter = 0
    while total_words <=55000:
        mod_lines = []
        for line in poem_lines:
          original_text = line
          tokens = original_text.split(' ')
          size = len(tokens)
          i = 0
          if line:
            while True:
              if i<size-1:
                new_text = tokens[:]
                new_text[i] = modify_string(new_text[i])
                new_text = ' '.join(new_text)
                modified_text = new_text.replace('[MASK]',unmasker(new_text,top_k=counter)[-1]['token_str'])
                tokens = modified_text.split(' ')
                size = len(tokens)
                i += 1
              else:
                break
            mod_lines.append(modified_text.strip(" ',./").replace(' ##',''))
          else:
            mod_lines.append('')
        oulipo = '\n'.join(mod_lines)+'\n\n'
        total_words+=len(oulipo.split())
        f.write(f'#############N={counter}#############\n\n')
        f.write(oulipo)
        print(counter,total_words)
        counter += 1
```
